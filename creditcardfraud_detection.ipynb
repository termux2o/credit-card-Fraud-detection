**GUI coding**

```python
from tkinter import *
import joblib

# Define a color palette dictionary for easy customization
color_palette = {
    "background": "#F0F0F0",  # Light gray
    "foreground": "#333333",  # Dark gray
    "button_normal": "#4CAF50",  # Green
    "button_hover": "#388E3C",  # Darker green (hover effect)
    "button_pressed": "#2E7D32",  # Even darker green (pressed effect)
    "label": "#000000",  # Black
    "entry": "#FFFFFF",  # White (for entry fields)
}


def show_entry_fields():
    values = [float(entry.get()) for entry in entries]  # Removed the non-breaking space here
    model = joblib.load("CREDIT_CARD_FRAUD_DETECT")
    pred = model.predict([values])
    result = "NORMAL TRANSACTION" if pred == 0 else "FRAUD TRANSACTION"
    result_label.config(text=result)

def on_enter(event):
    current_entry = event.widget
    index = entries.index(current_entry)
    next_index = (index + 1) % len(entries)
    entries[next_index].focus()

master = Tk()
master.title("Credit Card Fraud Detection Model")

# Set background color of the main window


label = Label(master, text="Credit Card Fraud Detection Model", fg=color_palette["foreground"], width=30)
label.grid(row=0, columnspan=2)

entries = []
labels = ['v1', 'v2', 'v3', 'v4', 'v5', 'v6', 'v7', 'v8', 'v9', 'v10',
          'v11', 'v12', 'v13', 'v14', 'v15', 'v16', 'v17', 'v18', 'v19', 'v20',
          'v21', 'v22', 'v23', 'v24', 'v25', 'v26', 'v27', 'v28', 'v29']

for i, label_text in enumerate(labels):
    label = Label(master, text=f'Enter value {label_text}', fg=color_palette["label"])  # Set label text color
    label.grid(row=i+1, column=0)
    entry = Entry(master, bg=color_palette["entry"])  # Set entry background color
    entry.grid(row=i+1, column=1)
    entry.bind('<Return>', on_enter)  # Bind Enter key
    entries.append(entry)

result_label = Label(master,text="Final prediction from model", fg=color_palette["label"])
result_label.grid(row=32, columnspan=2)

predict_button = Button(master, text="Predict", command=show_entry_fields, bg=color_palette["button_normal"], fg=color_palette["foreground"])
predict_button.grid(row=30, columnspan=2)
predict_button.bind("<Enter>", lambda e: predict_button.config(bg=color_palette["button_hover"]))
predict_button.bind("<Leave>", lambda e: predict_button.config(bg=color_palette["button_normal"]))  # Add hover effect

mainloop()
```
